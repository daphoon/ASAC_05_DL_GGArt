#캡셔닝 추출

```python
def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')

directory_path = r"C:\Users\NT551XCJ\Desktop\DLproject\images"  # 여기에 디렉토리의 올바른 경로를 입력하세요

# 지원되는 이미지 파일 확장자 목록을 작성합니다
image_extensions = ('.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff','webp')

caption1 = pd.read_csv("images_caption.csv", encoding='utf-8')
processed_files = set(caption1['filename'])

cap_list1=[]
err_list=[]
for filename in os.listdir(directory_path):
    if filename in processed_files:
        continue  # 이미 처리된 파일은 건너뜁니다
    
    try:
      # 파일 확장자가 이미지 파일 목록에 있는지 확인합니다
      image_path = os.path.join(directory_path, filename)

      base64_image = encode_image(image_path)

      headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {api_key}"
      }

      payload = {
        "model": "gpt-4o",
        "messages": [
          {
            "role": "user",
            "content": [
              {
                "type": "text",
                "text": """You're going to create a training dataset to train an image generating AI. The dataset is all images of characters from a game.
                    This AI will help you create a character with the desired design. I will give you an image and you will write a caption for it. The caption should follow the rules below.
                    - Describe the character's physical appearance in the image I provide.
                    - Describe the mood and feel of the character"""
              },
              {
                "type": "image_url",
                "image_url": {
                  "url": f"data:image/jpeg;base64,{base64_image}"
                }
              }
            ]
          }
        ],
        "max_tokens": 500
      }

      response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)

      res = response.json()
      if 'choices' in res:
          print(f"{filename} 정상")
          cap_list1.append([filename, res['choices'][0]['message']['content']])
    except:
            print(f"{filename}에 대한 오류: {res}")
            err_list.append(filename)
            continue


caption_data1=pd.DataFrame(
      data=cap_list1,
      columns=["filename","caption"]
  )

caption_data1.to_csv("images_caption.csv",encoding='cp949',index=False)

# 에러 이미지 저장
error_data=pd.DataFrame(
      data=error_list,
      columns=["filename"]
  )
error_data.to_csv("images_error.csv",encoding='',index=False)
```
