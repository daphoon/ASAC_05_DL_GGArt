```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import os
import requests

# 웹드라이버 설정
driver = webdriver.Chrome()

# 폴더 경로 설정 (이미지를 저장할 폴더 경로)
save_folder = 'pokemon_images'
os.makedirs(save_folder, exist_ok=True)

# 포켓몬 이미지 다운로드 함수
def download_pokemon_image(pokedex_number, save_number):
    # 사이트 접속
    url = f'https://www.pokemonkorea.co.kr/pokedex/view/{pokedex_number}?word=&characters=&area=&snumber=1&snumber2=1025&typetextcs=_________________&sortselval=number%20asc,number_count%20asc'
    driver.get(url)
    
    time.sleep(3)  # 페이지 로드 대기

    try:
        # 큰 이미지 요소 찾기
        large_image_element = driver.find_element(By.XPATH, '//*[@id="top"]/div[2]/div/div/div[1]/img')
        image_url = large_image_element.get_attribute('src')

        # 이미지 저장
        image_response = requests.get(image_url)
        image_path = os.path.join(save_folder, f'{save_number}.png')
        with open(image_path, 'wb') as file:
            file.write(image_response.content)
        print(f'Saved: {image_path}')
    except Exception as e:
        print(f'Error with pokedex_{pokedex_number}: {e}')

# 1부터 순서대로 포켓몬 이미지 다운로드 (파일명은 125번부터 시작)
start_save_number = 125
for i in range(1, 1026):  # 숫자 범위는 필요에 따라 조정
    download_pokemon_image(i, start_save_number)
    start_save_number += 1

# 드라이버 종료
driver.quit()
```
