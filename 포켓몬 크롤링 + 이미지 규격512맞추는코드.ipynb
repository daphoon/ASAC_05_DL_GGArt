```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import os
import requests

# 웹드라이버 설정
driver = webdriver.Chrome()

# 폴더 경로 설정 (이미지를 저장할 폴더 경로)
save_folder = 'pokemon_images'
os.makedirs(save_folder, exist_ok=True)

# 포켓몬 이미지 다운로드 함수
def download_pokemon_image(pokedex_number, save_number):
    # 사이트 접속
    url = f'https://www.pokemonkorea.co.kr/pokedex/view/{pokedex_number}?word=&characters=&area=&snumber=1&snumber2=1025&typetextcs=_________________&sortselval=number%20asc,number_count%20asc'
    driver.get(url)
    
    time.sleep(3)  # 페이지 로드 대기

    try:
        # 큰 이미지 요소 찾기
        large_image_element = driver.find_element(By.XPATH, '//*[@id="top"]/div[2]/div/div/div[1]/img')
        image_url = large_image_element.get_attribute('src')

        # 이미지 저장
        image_response = requests.get(image_url)
        image_path = os.path.join(save_folder, f'{save_number}.png')
        with open(image_path, 'wb') as file:
            file.write(image_response.content)
        print(f'Saved: {image_path}')
    except Exception as e:
        print(f'Error with pokedex_{pokedex_number}: {e}')

# 1부터 순서대로 포켓몬 이미지 다운로드 (파일명은 125번부터 시작)
start_save_number = 125
for i in range(1, 1026):  # 숫자 범위는 필요에 따라 조정
    download_pokemon_image(i, start_save_number)
    start_save_number += 1

# 드라이버 종료
driver.quit()
```

```python
# image.resize 함수 사용 
# 그냥 단순히 지정한 크기에 맞춰서 사이즈조정하는거임 

from PIL import Image
import os

# 폴더 경로 설정 (이미지가 저장된 폴더 경로)
image_folder = 'pokemon_images'
resized_folder = '포켓몬_최종1'
os.makedirs(resized_folder, exist_ok=True)

# 이미지 크기 조정 함수
def resize_image(image_path, save_path, target_size=512):
    try:
        image = Image.open(image_path)
        width, height = image.size
        
        # 비율 유지하며 크기 조정
        if width > height:
            new_width = target_size
            new_height = int((target_size / width) * height)
        else:
            new_height = target_size
            new_width = int((target_size / height) * width)
        
        resized_image = image.resize((new_width, new_height), Image.LANCZOS)
        resized_image.save(save_path)
        print(f'Resized and saved: {save_path}')
    except Exception as e:
        print(f'Error resizing {image_path}: {e}')

# 이미지 폴더의 모든 이미지 파일 순회
for filename in os.listdir(image_folder):
    if filename.endswith('.png'):
        image_path = os.path.join(image_folder, filename)
        save_path = os.path.join(resized_folder, filename)
        resize_image(image_path, save_path)

print("All images have been resized.")
```

```python
# image.thumbnail함수 사용 
# 종횡비 유지, 원본 이미지를 초과하지않으며 이미지 확대하지않음 

from PIL import Image
import os

# 폴더 경로 설정 (이미지가 저장된 폴더 경로)
image_folder = '포트나이트'
resized_folder = '포트나이트_시도2'
os.makedirs(resized_folder, exist_ok=True)

# 이미지 크기 조정 함수
def resize_image(image_path, save_path, target_size=512):
    try:
        image = Image.open(image_path)
        
        image.thumbnail((target_size, target_size), Image.LANCZOS)
        image.save(save_path, quality=95)
        print(f'Resized and saved: {save_path}')
    except Exception as e:
        print(f'Error resizing {image_path}: {e}')

# 이미지 폴더의 모든 이미지 파일 순회
for filename in os.listdir(image_folder):
    if filename.endswith('.png'):
        image_path = os.path.join(image_folder, filename)
        save_path = os.path.join(resized_folder, filename)
        resize_image(image_path, save_path)

print("All images have been resized.")
```
